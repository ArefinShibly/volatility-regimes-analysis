```python
!pip install yfinance pandas

import yfinance as yf
import pandas as pd

# Example: GOLD Future weekly data (2015-2024)
ticker = 'GC=F'
data = yf.download(ticker, start='2015-01-01', end='2024-12-31', interval='1wk')

# Save as CSV
data.to_csv('/content/GOLDFUTURE_weekly_2015_2024.csv')
print("✅ Data downloaded successfully")
```
